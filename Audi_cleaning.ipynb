```python
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport
```

```python
df=pd.read_csv("audi.csv")
```

```python
profile = ProfileReport(df, title="Report")
profile
```

```python
df.isnull().sum() 
```

```python
#11 rows are filled with completely  nan values
df = df[df['Marke'].notna()]
df=df.drop_duplicates()
```

```python
df.isnull().sum() 
```

```python
del df["Marke"]
```

```python
#From report: HU bis has 2922 (22.3%) missing values	Missing
#Umweltplakette has 2415 (18.5%) missing values	Missing
#Schadstoffklasse has 2484 (19.0%) missing values
df = df[df['HU bis'].notna()]
df = df[df["Schadstoffklasse"].notna()]
df = df[df["Umweltplakette"].notna()]
```

```python
df.isnull().sum()
```

```python
#other missing values would be filled
df.info()
```

```python
df=df.fillna(method='ffill')
```

```python
#month is important
df["HU bis"].replace({"2021": "06 2021", "2022": "01 2022", "2023":"01 2023"}, inplace=True)
```

```python
df["HU bis"]=df["HU bis"].replace(regex={r'März': '03', 'Mai': '05', "Januar":"01",
                                         "Februar": "02", "April":"04", "Juni":"06",
                                         "Juli":"07","August":"08","September":"09",
                                         "Oktober":"10","November":"11","Dezember":"12"})
```

```python
df["Erstzulassung"]=df["Erstzulassung"].replace(regex={r'März': '03', 'Mai': '05', "Januar":"01",
                                         "Februar": "02", "April":"04", "Juni":"06",
                                         "Juli":"07","August":"08","September":"09",
                                         "Oktober":"10","November":"11","Dezember":"12"})
```

```python
df.nunique()
```

```python
df.Getriebe.unique()
```

```python
df["Getriebe"].replace({'Automatik':"Automatic", 'Manuell':"Manuel"}, inplace=True)
```

```python
df.Fahrzeugzustand.unique() 
```

```python
df["Fahrzeugzustand"].replace({'Beschädigtes Fahrzeug':"damaged", 'Unbeschädigtes Fahrzeug':"non-damaged"}, inplace=True)
```

```python
df.head()
```

```python
(df.Erstzulassung[df.Erstzulassung.str.len() ==4]).unique()
```

```python
df["Erstzulassung"]=df["Erstzulassung"].replace({'1997':"01 1997", '2002':"01 2002", '1994':"01 1994", '2007':"01 2007",
                      '1991':"01 1991", '1993':"01 1993", '2010':"01 2010", '1995':"01 1995",
       '1992':"01 1992", '2003': "01 2003", '2008':"01 2008", '2011':"01 2011", '2012':"01 2012", 
       '2001':"01 2001", '1996': "01 1996", '2019':"01 2019",
       '1998':"01 1998", '1999':"01 1999", '1990': "01 1990", '2006':"01 2006", '2009':"01 2009", 
       '2013':"01 2013", '2004':"01 2004", '2014':"01 2014",
       '2000':"01 2000", '2015': "01 2015", '2018':"01 2018", '2016':"01 2016", '2017':"01 2017", '2005':"01 2005"})
```

```python
df[['reg_month', 'reg_year']] = df['Erstzulassung'].str.split(' ', 1, expand=True)
```

```python
df[['inspection_month', 'inspection_year']] = df['HU bis'].str.split(' ', 1, expand=True)
```

```python
del df['Erstzulassung']
del df['HU bis']
```

```python
df["Fahrzeugtyp"].unique()
```

```python
df["Fahrzeugtyp"].replace({'SUV/Geländewagen':"suv", 'Kleinwagen':"small car","Kombi":"Wagon",'Andere Fahrzeugtypen':"other"}, inplace=True)
```

```python
df["Kraftstoffart"].unique()
```

```python
df["Kraftstoffart"].replace({'Erdgas (CNG)':"CNG",'Autogas (LPG)':"LPG",'Andere Kraftstoffarten':"other"}, inplace=True)
```

```python
df["Außenfarbe"].unique()
```

```python
df["Außenfarbe"].replace({'Schwarz':"black", 'Grau':"grey", 'Braun':"brown", 'Silber':"silver", 'Weiß':"white", 'Blau':"blue", 'Grün':"green",
       'Rot':"red", 'Andere Außenfarben':"other",'Gelb':"yellow"}, inplace=True)
```

```python
df["Material Innenausstattung"].unique()
```

```python
df["Material Innenausstattung"].replace({'Vollleder':"full-leather", 'Stoff':"fabric", 'Teilleder':"part-leather", 
       'Andere Materialien Innenausstattung':"other"}, inplace=True)
```

```python
df["Kilometerstand"]  = df["Kilometerstand"] .str.replace("km","")
df["Kilometerstand"]  = df["Kilometerstand"] .str.replace(",","")
df["Leistung"]  = df["Leistung"] .str.replace("PS","")
```
